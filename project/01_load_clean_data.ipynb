````python
if update_files is True: 
    # Set up DATA DIR
    print "Creating data dir directory at {}".format(DATA_DIR)
    if not os.path.isdir(DATA_DIR): 
        !mkdir $DATA_DIR
    else:
        # delete all contents of destination directory! 
        #!rm -rf $DATA_DIR/*
        pass

    # copy all data files in SOURCE_DATA_DIR to DATA_DIR
    print "Copying files...".format(DATA_DIR)
    !cp $SOURCE_DATA_DIR/*.json* $DATA_DIR/
    
    scriptfilename = os.path.join(DATA_DIR, 'build_all_derived_data.sh')

    print "Creating bash shell script: {}".format(scriptfilename)
    # Create preprocessing script
    with open(scriptfilename, 'w+') as sf:
        # Change working directory
        sf.write('cd ${0%/*}' + '\n')
        sf.write('echo Executing script at : $PWD' + '\n')
        
        split_files = !ls $SOURCE_DATA_DIR | grep '.json_'
        split_roots = list(set([f.split('.')[0] for f in split_files]))
        
        # RAW Data: keep only desired records, records for selected user
        # 1. add newlines for each record
        # 2. keep only lines with the desired user
        # 3. third stitch together user raw files
        sf.write('\necho "Trimming raw json files for user {}..."'.format(selected_user) + '\n')
        for raw_file in [f for f in split_files if f.startswith('raw')]:
            raw_user_filename = "raw_user_{}.json_{}".format(
                selected_user.lstrip("#"),
                raw_file.split('_')[-1]
                )
            sf.write(
            '''sed "s/{\\"@type/\\n{\\"@type/g" %(raw_file)s | grep \"^{.*%(userid)s.*},$\" > %(raw_user_filename)s''' % {
                'raw_file': raw_file,
                'userid': selected_user,
                'raw_user_filename' : raw_user_filename,
                } + '\n'
            )
        
        selected_user_root = 'raw_user_{}'.format(selected_user.lstrip("#"))
        selected_user_raw_filename = '{}.json'.format(selected_user_root)
        
        split_roots.append(selected_user_root)
        
        # Concat split files 
        #  not raw.json because it's huge. 
        sf.write('\necho Concatenating split JSON files...' + '\n')
        for root in [r for r in split_roots if r!='raw']:
            sf.write('cat {}.json_* > {}.json'.format(root, root) + '\n')
        
        # delete split files already concat'ed
        sf.write('\necho Deleting split files...' + '\n')
        for root in [r for r in split_roots if r!='raw']:
            sf.write('#rm -rf {}.json_* > {}.json'.format(root, root) + '\n')
        
        # SED HACKS
        # SED JSON files to make appropriate record format
        sf.write('\necho Executing sed stich hack fix...' + '\n')
        # 1. Apply to all *.json especially raw.json 
        sf.write("""find *.json -type f -exec sed -i -r -e 's/^\{"records"://' -e 's/}$//' {} \;""" + '\n')
        # 2. Apply to specifically raw_user_*.json
        # Add '[' at start and end
        #sf.write("""#find raw_user_*.json -type f -exec sed -i -r '1s/^/{\"records\":[/' {} \;""" + '\n')
        sf.write("""find raw_user_*.json -type f -exec sed -i '1s/^{/[{/' {} \;""" +'\n')
        sf.write("""find raw_user_*.json -type f -exec sed -i '$ s/},$/}]/' {} \;""" +'\n')

        
        # CREATE UPDATED FILE
        sf.write ('touch {}'.format(UPDATED_FILE_TRACKER))
        
        sf.write('\necho "...done!"')
    
    # Run preprocessing script
    !bash $scriptfilename
else: 
    print "Not executing script"
````

````python
"""Reindex and prep the CMD data
"""

idf = cmd_df[cmd_df.user == user]

# Reindex, tseries with timezone conversions
idf = idf.dropna(subset=['localTime'])
idf.localTime = pd.DatetimeIndex(idf.localTime, tz=tz).tz_convert('UTC')

idf.set_index('localTime', drop=False, inplace=True)
````

### Add Sleep periods (computed from CMD sleepDuration)

````python
""" 
Obtain sleep ranges
 and add to idf_cov_list
"""
# define lambda function
sleepstart = lambda ss: ss['localTime'] - timedelta(seconds=ss['value']*3600)

isleepdf =  idf[idf['@class'] == 'sleepDuration']
if len(isleepdf) <= 0:
    print "Warning: No Sleep Periods found for selected user / time range."
else: 
    isleepdf['sleepstart'] = isleepdf.apply(sleepstart, axis=1)

    # Sleep columns
    keep_cols =    ['updated',   'sleepstart',  'value', ]
    renamed_cols = ['sleep_end', 'sleep_start', 'sleep_duration']

    # Manipulate columns
    isleepdf = isleepdf[keep_cols]
    isleepdf.columns = renamed_cols

    # Save
    idf_cov_list.append(isleepdf)
````

### Add CMD: HeartRate 

````python
"""
For the selected cov
 - subset
 - resample
 - add data to our df
"""
# INPUT select covariate values
cov = 'heartRate'
cov_how_resample = 'mean'
resample_freq = 'T'

keep_cols =    ['value'   ]
renamed_cols = ['heart_rate']

assert cov in cmd_df['@class'].unique(), "ERROR: '{}' data does not exist.".format(cov)

# -- Subsetting --
idf_cov = idf[idf['@class'] == cov]

# Resampling
idf_cov = idf_cov.resample(resample_freq, how=cov_how_resample)

# Manipulate columns
idf_cov = idf_cov[keep_cols]
idf_cov.columns = renamed_cols

# Save
idf_cov_list.append(idf_cov)
````

## Data Completeness by minimum frequency

````python
user = '#11:392'
min_freq = 'D'
min_count = 1

freq_dict = {
    #u'group'           : 'D',
    #u'config'          : 'D',
    #u'contact'         : 'D',
    #u'patch'           : 'D',
    #u'term'            : 'D',
    #u'tag'             : 'D',
    #u'patchStatus'     : 'D',
    u'bodyAngle'       : '5T',
    u'heartRate'       : 'T',
    u'stepCount'       : 'H',
    u'sleepDuration'   : 'D',
    #u'bodyWeight'      : 'D' ,
    u'energyExpend'    : 'H',
    u'skinTemp'        : '5T',
    u'sleepQuality'    : 'D', 
    u'fatigue'         : 'D',
    u'soreness'        : 'D',
    u'recovery'        : 'D',
    }

by_freq = []
for cov, min_freq in freq_dict.iteritems():
    #print cov, min_freq

    tt = cmd_df[(cmd_df.user == user) & (cmd_df['@class'] == cov)]\
            .set_index('localTime')\
            .truncate(before=first_date, after=last_date)\
            .value\
            .resample(min_freq, how='count') >= min_count
    tt.name = cov
    
    if min_freq == 'H' : tt = tt.resample('D', how='count') >= 24
    if min_freq == 'T' : tt = tt.resample('D', how='count') >= 24 * 60
    if min_freq == '5T': tt = tt.resample('D', how='count') >= 24 * (60/5)
    
    by_freq.append(tt)

freq_df = pd.DataFrame(by_freq).T
````
