###8 & 9. What is the average age of passengers who survived vs those who did not survive?

```python
titanic.groupby('Survived').Age.agg(['mean', 'std'])
```

###10. At this (early) point in our analysis, what might you infer about any patterns you are seeing?

Women were much more likely to survive, 74% vs 19% and while the average age was 28 vs 30 for those who didn't survived vs those who did, the std is pretty large indicating that a greater range of ages 16-48 (68% of data) survived.
