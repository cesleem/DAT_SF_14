```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
from sklearn import neighbors, datasets, feature_selection
from sklearn.cross_validation import train_test_split, cross_val_score
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
clf = neighbors.KNeighborsClassifier(10)
scores = cross_val_score(clf, iris_df.values, iris.target, cv = 5)
np.mean(scores)
```

```python
mean_scores = []
for i in range(1,len(iris.data) / 3):
    clf = neighbors.KNeighborsClassifier(i, weights='uniform')
    clf.fit(iris.data, iris.target)
    cv_scores = cross_val_score(clf, iris_df.values, iris.target, cv = 5)
    mean_scores.append(np.mean(cv_scores))
```

```python
def cross_validation(X, y, clf, cv = 5, random_seed = 1):
    np.random.seed(seed=random_seed)
    Xy = np.append(X, y.reshape(-1, 1), axis = 1)
    np.random.shuffle(Xy)
    Xy_split = np.array_split(Xy, cv, axis = 0)
    cv_scores = []
    for i in range(cv):
        x_list = range(cv)
        x_list.remove(i)
        test_cv = Xy_split[i]
        train_cv = np.array([], dtype= np.float64).reshape(0,Xy.shape[1])
        for i in x_list:
            train_cv = np.vstack([train_cv,Xy_split[i]])

        # split columnwise
        X_train, y_train = np.hsplit(train_cv, np.array([Xy.shape[1] - 1]))
        X_test, y_test = np.hsplit(test_cv, np.array([Xy.shape[1] - 1]))

        # reshaping y_train from 2D to 1D
        y_train = y_train[:,0]

        # fit to model
        clf.fit(X_train, y_train)

        # score train-test-splits
        cv_scores.append(clf.score(X_test, y_test))
    return cv_scores
```
